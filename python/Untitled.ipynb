```python

import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
from pypalettes import load_cmap

# Load your colormap and define colors
cmap = load_cmap("DeathValley")
N = 10  # number of colors in the cycle
colors = cmap(np.linspace(0, 1, N))

# Use colors in Matplotlib
plt.rcParams['axes.prop_cycle'] = cycler(color=colors)

# ---- Print colors with RGB values visually ----
fig, ax = plt.subplots(figsize=(6, 0.5 * N))

for i, color in enumerate(colors):
    r, g, b, a = color  # unpack RGBA
    r = int(r*255)
    g = int(g*255)
    b = int(b*255)
    ax.add_patch(plt.Rectangle((0, i), 1, 1, color=color))
    ax.text(1.05, i + 0.5, f"RGB = ({r:.3f}, {g:.3f}, {b:.3f})",
            va='center', fontsize=10)

ax.set_xlim(0, 3)
ax.set_ylim(0, N)
ax.axis('off')

plt.tight_layout()
plt.show()

```
